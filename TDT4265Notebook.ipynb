```python
# go trough all the label txt files in data/combined and edit the  txt file by removing any line that does not start with 0 1 2 3
# this is done because the model only predicts 4 classes and we have 5 classes in the dataset

import os
#firt set path to the folder with the label files
path = r"E:\Lofty\Documents\TDT4265\Data\Combined\test\labels"

#then go through all the files in the folder
for filename in os.listdir(path):
    print(filename)

    #open the file
    with open(os.path.join(path, filename), "r") as f:
        #read the lines in the file
        lines = f.readlines()
    #open the file again in write mode
    with open(os.path.join(path, filename), "w") as f:
        #go through all the lines in the file
        for line in lines:
            #if the line starts with 0 1 2 or 3 then write it to the file
            if line[0] == "0" or line[0] == "1" or line[0] == "2" or line[0] == "3":
                f.write(line)
```
